## Problem #1

Show that if $P$ is an orthogonal projector, then $I - 2P$ is unitary.  Prove this algebraically, and give a geometric interpretation.


#### Solution
$$~~~$$
If $P$ is an orthogonal projector, then $P = P^*$.

Let $A = I−2P $. For $A$ to be unitary, then $A^* = A^{-1}$ therefore to show that $A$ is unitary then we must show that $A^*A = I$ and this is as below;
\begin{align}
A^*A &= (I−2P)^*(I−2P)\\
     & = (I^* -2P^*)(I−2P)\\
     & = (I -2P^*)(I−2P)\\
     & = I - 2P- 2P^* + 4P^*P\\
\end{align}
But $P = P^*$;
\begin{align}
\Rightarrow A^*A & = I - 4P + 4P^2\\
&= I - 4P + 4P\\
&=I
\end{align}
Therefore $I−2P$ is unitary since $A^*A = (I−2P)^*(I−2P) = I$.

$\textbf {Geometric interpretation}$


Suppose vector $\mathbf w\in range(P)$ and we have that;







\begin{align}
(I-2P)\mathbf w &= \mathbf w - 2P\mathbf w\\
    & = (I-P)\mathbf w - P\mathbf w
\end{align}

Therefore the geometric interpretation of this is that $(I-2P)\mathbf w$ is the reflection of vector $\mathbf w$ along the space orthogonal to $P\mathbf w$

## Problem #2

Given $A \in \mathcal C^{m \times n}$ with $m \ge n$, show that $A^*A$ is non-singular if and only if $A$ has full rank. 


#### Solution

Let $C = A^*A$  be a non-singular matrix. This means that $ det(C) = det(A^*A)\neq 0$ implying that matrix $C$ has full rank.
The adjoint of $C$ is then given by;
\begin{align}
C^* &= (A^*A)^* = A^*A\\
\Rightarrow det(C^*) &= det(A^*A)\neq 0
\end{align}
This then implies that $C^*$ also has full rank.

Since $A^*A$ is hermitian and $det(A^*A)\neq 0$, then $det(A)\neq 0$ thus $A$ has full rank.

Therefore $A^*A$ is non singular if and only if $A$ has full rank.

## Problem #3 

The singular vectors and values of a matrix $A$ satisfy

\begin{equation}
A \mathbf v_i = \sigma_i \mathbf u_i
\end{equation}

When we write this in matrix form, we obtain the SVD of $A$, given by 

\begin{equation}
A = U \Sigma V^*
\end{equation}

By convention, we order the singular values in the diagonal of $\Sigma$ in decreasing order so that 

\begin{equation}
\Sigma_{11} \ge \Sigma_{22} \ge \Sigma_{33} \ge \dots \ge \Sigma_{rr} > 0
\end{equation}

where $r$ is the rank of $A$.  Show that it in fact doesn't matter how we arrange the singular vectors into columns of $U$ and $V$ and $\Sigma$, as long as we are consistent with the ordering of the vectors within $U$ and $V$. 


#### Solution
To show that it doesn't matter how we arrange the singular vectors into columns of $U$ and $V$ with their correponding singular values in $\Sigma$ as long as we are consistent, we can use a permutation matrix say $P$ to show this.

Assuming $A$ is an $m\times n$ matrix with $m\geq n$ and that it has full rank that is rank(A) = $r$ = $n$, $P$ will then be a $n\times n$ matrix with exactly one entry of 1 in each row and each column and 0s elsewhere. 

Let
\begin{align}
U = \begin{pmatrix}u_{11}&u_{12}&\cdots& u_{1r}\\u_{21}&u_{22}&\cdots& u_{2r}\\ \vdots&\vdots&\vdots&\vdots \\u_{m1}&u_{m2}&\cdots& u_{mr} \end{pmatrix}, ~~~~~ \Sigma = \begin{pmatrix}\Sigma _{11}&0&\cdots& 0\\0&\Sigma _{22}&\cdots& 0\\\vdots&\vdots&\ddots&\vdots\\0&0&\cdots&\Sigma _{rr}\end{pmatrix}~~~~~~~ V = \begin{pmatrix}v_{11}&v_{12}&\cdots& v_{1r}\\v_{21}&v_{22}&\cdots& v_{2r}\\ \vdots&\vdots&\vdots&\vdots \\v_{r1}&v_{r2}&\cdots& v_{rr} \end{pmatrix}
\end{align}
$$~~~$$
If I want to interchange $\Sigma _{11}$ with $\Sigma _{22}$, then $P = \begin{pmatrix}0&1&0&\cdots& 0\\1&0&0&\cdots& 0\\0&0&1&\cdots& \\\vdots&\vdots&&\ddots&\vdots\\0&0&0&\cdots&1\end{pmatrix} $

The diagonal elements $\Sigma _{11}$ and $\Sigma _{22}$ can be interchaged a shown below;


\begin{align}
 \begin{pmatrix}0&1&0&\cdots& 0\\1&0&0&\cdots& 0\\0&0&1&\cdots& \\\vdots&\vdots&&\ddots&\vdots\\0&0&0&\cdots&1\end{pmatrix}\begin{pmatrix}\Sigma _{11}&0&\cdots& 0\\0&\Sigma _{22}&\cdots& 0\\\vdots&\vdots&\ddots&\vdots\\0&0&\cdots&\Sigma _{rr}\end{pmatrix} \begin{pmatrix}0&1&0&\cdots& 0\\1&0&0&\cdots& 0\\0&0&1&\cdots& \\\vdots&\vdots&&\ddots&\vdots\\0&0&0&\cdots&1\end{pmatrix}
&= \begin{pmatrix}\Sigma _{22}&0&\cdots& 0\\0&\Sigma _{11}&\cdots& 0\\\vdots&\vdots&\ddots&\vdots\\0&0&\cdots&\Sigma _{rr}\end{pmatrix}\\
P~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~ \Sigma~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~ P ~~~~~~~~ ~~~~~~ &
\end{align} 
$$~~~$$
Since consistence is a key factor we as well have to interchage the columns of $U$ with respect to $\Sigma _{11}$ and $\Sigma _{22}$ as shown below;

\begin{align}
  \begin{pmatrix}u_{11}&u_{12}&\cdots& u_{1r}\\u_{21}&u_{22}&\cdots& u_{2r}\\ \vdots&\vdots&\vdots&\vdots \\u_{m1}&u_{m2}&\cdots& u_{mr} \end{pmatrix}\begin{pmatrix}0&1&0&\cdots& 0\\1&0&0&\cdots& 0\\0&0&1&\cdots& \\\vdots&\vdots&&\ddots&\vdots\\0&0&0&\cdots&1\end{pmatrix}
&= \begin{pmatrix}u_{12}&u_{11}&\cdots& u_{1r}\\u_{22}&u_{21}&\cdots& u_{2r}\\ \vdots&\vdots&\vdots&\vdots \\u_{m2}&u_{m1}&\cdots& u_{mr} \end{pmatrix}\\
U~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~ P~~~~~~~~~~ &
\end{align}

Lastly we have to interchage the rows of $V^*$ with respect to $\Sigma _{11}$ and $\Sigma _{22}$ as shown below


\begin{align}
  \begin{pmatrix}0&1&0&\cdots& 0\\1&0&0&\cdots& 0\\0&0&1&\cdots& \\\vdots&\vdots&&\ddots&\vdots\\0&0&0&\cdots&1\end{pmatrix}
\begin{pmatrix}\bar{v}_{11}&\bar{v}_{21}&\cdots& \bar{v}_{r1}\\\bar{v}_{12}&\bar{v}_{22}&\cdots& \bar{v}_{r2}\\ \vdots&\vdots&\vdots&\vdots \\\bar{v}_{1r}&\bar{v}_{2r}&\cdots& \bar{v}_{rr} \end{pmatrix}& = \begin{pmatrix}\bar{v}_{12}&\bar{v}_{22}&\cdots& \bar{v}_{r2}\\\bar{v}_{11}&v_{21}&\cdots& \bar{v}_{r1}\\ \vdots&\vdots&\vdots&\vdots \\\bar{v}_{1r}&\bar{v}_{2r}&\cdots& \bar{v}_{rr} \end{pmatrix}\\
P~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~ V^*~~~~~~~~~~~~~~~~~~~  &
\end{align}

We notice then that to re-arrange the singular vectors into columns of $U$ and $V$ with their correponding singular values in $\Sigma$ consistently $A$ becomes;

\begin{align}
  A &=  (U P) (P \Sigma P) (P V^*)\\
  & = U P^2 \Sigma P^2  V^*
\end{align}

But $P^2 = I$, $\Rightarrow A= U  \Sigma   V^*$. This therefore shows that it doesn't matter how we arrange the singular vectors into columns of $U$ and $V$ with their correponding singular values in $\Sigma$ as long as we are consistent.

## Problem #4

Let $P \in \mathcal C^{m \times m}$ be a non-zero projector.  Show that $\Vert P \Vert_2 \ge 1$ with equality if and only if $P$ is an orthogonal projector.  

**Hint:** Use the definition of a project presented in the introduction to Lecture 6. 

#### Solution
Since $P$ is a non-zero projector, then $\left \Vert P \right \Vert _2 \neq 0 $ and $P^2 = P$.

Suppose  $\mathbf v \in  range(P)$ and  $\mathbf v = P\mathbf x$ for some $\mathbf x$, then $\mathbf v
$lies exactly on its own shadow, and applying the projector on $ \mathbf v$ results in;
\begin{align}
P\mathbf v = P(P \mathbf x) = P^2 \mathbf x = P\mathbf x = \mathbf v
\end{align}
since 
\begin{align}
P\mathbf v &= \mathbf v\\
\Rightarrow \left \Vert P \mathbf v \right \Vert _2 &= \left \Vert  \mathbf v  \right \Vert _2
\end{align}



Using the Cauchy–Schwarz inequality, $\left \Vert P \mathbf v \right \Vert _2  \leq \left \Vert P \right \Vert _2 \left \Vert \mathbf v \right \Vert _2 $ the above equation becomes;

\begin{align}
\left \Vert P \right \Vert _2 \left \Vert \mathbf v \right \Vert _2 &\geq \left \Vert \mathbf v \right \Vert _2\\
\Rightarrow \left \Vert P \right \Vert _2 & \geq 1
\end{align}

If P is an orthogonal projector, then $P = P ^*$. 

Let $P =U\Sigma V^*$ be the SVD  form of $P$ where $U$ and $V$ are the left and right singular vectors respectively, $UU^* =VV^* =I$ and $\Sigma $ is a diagonal matrix of singular values $\sigma _1\geq\sigma _2\geq ... \geq \sigma _m >0 $.

From Lecture 3, we know that  $\left \Vert P \right \Vert _2 = \max _\limits{{1\leq i \leq m}} \sigma _i = \sigma _1$. 

Therefore $\left \Vert P \right \Vert _2 = 1$ if and only if $\sigma _1 = 1$ and $\sigma _1 = 1$ iff $\Sigma = I$.

We also know that;
\begin{align}
 P^2 &= P\\
 \Rightarrow U\Sigma V^* U\Sigma V^* &= U\Sigma V^*\\
 V^* U\Sigma &= I
\end{align}
This means that $\Sigma = I$ iff $V^* U = I$ i.e. $V=U$ which also means that $\Sigma = I$ if and only iff P is an orthogonal progector.

Therefore $\left \Vert P \right \Vert _2 = 1$ if and only if P is an orthogonal progector.

## Problem #6

Let $U$ be the matrix of singular vectors of the matrix $A$.  Show that $P = UU^T$ is an orthogonal projector onto $\mbox{range}(A)$. Assume that the columns of $A$ are linearly independent.



#### Solution


If we assume that the columns of matrix $A$ are linearly independent and that $m\geq n$, then $range(A)$ is spanned by all columns of $A$.
Thus $A$ is full rank with $rank(A) =n$.

Let $A = U\Sigma V^T$ be the reduced SVD form of an $m\times n$ matrix $A$ where $U$ is an $m\times n$ matrix of orthonormal left singular vectors
,$V$ is an $n\times n$ matrix of orthonormal right singular vectors and $\Sigma$ is $n\times n$ diagonal matrix of singular values. We then have that $V^TV = I$,  and $\Sigma ^T = \Sigma $ 



If P is an orthogonal projection onto $range(A)$, then it is given by;
\begin{align}
P &= A(A^TA)^{-1}A^T\\
& = U\Sigma V^T [(U\Sigma V^T)^T U\Sigma V^T]^{-1} (U\Sigma V^T)^T\\
& = U\Sigma V^T [V\Sigma U^T U\Sigma V^T]^{-1} V \Sigma U^T\\
& = U\Sigma V^T [V\Sigma^2 V^T]^{-1} V \Sigma U^T\\
&= U\Sigma V^T V^{-T}\Sigma^{-2} V^{-1} V \Sigma U^T\\
&= U\Sigma V^T V \Sigma^{-2} V^T V \Sigma U^T\\
&= U\Sigma \Sigma^{-2}  \Sigma U^T\\
\therefore P& = U U^T
\end{align}
