## Problem 1 (TB 2.1)
<hr style="border-width:3px; border-color:black"></hr>


Exercise 2.1 (Lecture 2), Trefethen and Bau, page 15.  

Show that if matrix $A$ is both triangular and unitary, then it is diagonal.  

#### Hint

Use a result from Exercise TB 1.3.  

#### Solution

If matrix $A$ is unitary, then $A^{-1} = A$.

From Exercise TB 1.3., we proved that if  matrix $A$ is upper triangular, then $A^{-1}$ is also upper triangular and this implies that 
if $A$ is an upper triangular unitary matrix, then its transpose conjugate $A^*$ is also upper triangular.

Similarly if $A$ is a lower triangular unitary matrix,then its transpose conjugate $A^*$ is also lower triangular.

Therefore, since both $A$ and $A^*$ are triangular then $A$ is a diagonal matrix.

## Problem 2 (TB 2.2)
<hr style="border-width:2px; border-color:black"></hr>

Exercise 2.2 (Lecture 2), Trefethen and Bau, page 15.

The Pythagorean theorem asserts that for a set of $n$ orthogonal vectors $\{\mathbf x_i\}$, 

\begin{equation}
\left\Vert \sum_{i=1}^n \mathbf x_i \right\Vert^2 = \sum_{i=1}^n \left\Vert \mathbf x_i \right\Vert^2
\end{equation}

For this problem (and all problems in Lecture 2, you may assume that $\Vert \cdot \Vert$ is the 2-norm.

**(a)** Prove this for $n = 2$ by direct computation. 

**(b)** Show this computational also establishes the general case by induction.



#### Solution

(use this space to answer this question). 

$\textbf {(a)}$

For $n = 2$, we have that;

\begin{align}
\left\Vert \sum_{i=1}^2 \mathbf x_i \right\Vert^2 &=  \left\Vert \mathbf x_1 + \mathbf x_2  \right\Vert^2\\
&= (\mathbf x_1+ \mathbf x_2)^* (\mathbf x_1+ \mathbf x_2)\\
&= \mathbf {x_1}^* \mathbf x_1 + \mathbf x_1^* \mathbf x_2+ \mathbf x_2^* \mathbf x_1+\mathbf x_2^* \mathbf x_2
\end{align}
$$~~~~~$$
But since $\mathbf x_i$ are othorgonal vectors, this implies that $\mathbf x_1^*\mathbf x_2= \mathbf x_2^*\mathbf x_1= 0$
$$~~~~~$$
\begin{align}
\therefore \left\Vert \sum_{i=1}^2 \mathbf x_i \right\Vert^2 &= \mathbf x_1^* \mathbf x_1 +\mathbf x_2^* \mathbf x_2\\
&= \left\Vert \mathbf x_1 \right\Vert^2 +\left\Vert \mathbf x_2 \right\Vert^2\\
&= \sum_{i=1}^2 \left\Vert \mathbf x_i \right\Vert^2
\end{align}

$\text {(b)}$

let n = k,
Showing for k-1, we have that;
\begin{align}
\left\Vert \sum_{i=1}^{k-1} \mathbf x_i \right\Vert^2 &=  \left\Vert \mathbf x_1 + \mathbf x_2  + ... + \mathbf x_{k-1}  \right\Vert^2 \\
&= (\mathbf x_1+ \mathbf x_2+ ... + \mathbf x_{k-1})^* (\mathbf x_1+ \mathbf x_2 + ... + \mathbf x_{k-1})\\
&= \mathbf x_1^* \mathbf x_1 + \mathbf x_1^* \mathbf x_2+ ... + \mathbf x_1^*\mathbf x_{k-1} + \mathbf x_2^*\mathbf x_1+\mathbf x_2^* \mathbf x_2+ ... +  \mathbf x_2^* \mathbf x_{k-1}+ ...+ \mathbf x_{k-1}^* \mathbf x_1 + \mathbf x_{k-1}^* \mathbf x_2+...+ \mathbf x_{k-1}^* \mathbf x_{k-1} 
\end{align}
$$~~~~~$$
But since $\mathbf x_i$ are othorgonal vectors, this implies that $\mathbf x_i^*\mathbf x_j= 0$ for $i\neq j$. We then get;
$$~~~~~$$
\begin{align}
\therefore \left\Vert \sum_{i=1}^{k-1} \mathbf x_i \right\Vert^2 &= \mathbf x_1^* \mathbf x_1 +\mathbf x_2^* \mathbf x_2 + ... + \mathbf x_{k-1}^* \mathbf x_{k-1} \\
&= \left\Vert \mathbf x_1 \right\Vert^2 +\left\Vert \mathbf x_2 \right\Vert^2+ ... + \left\Vert \mathbf x_{k-1} \right\Vert^2\\
&= \sum_{i=1}^{k-1} \left\Vert \mathbf x_i \right\Vert^2
\end{align}
Since the computation holds for $n = k-1$, then by induction the Pythagorean theorem holds $\forall n\in \mathbb{N}$

## Problem 3  (TB 2.3)
<hr style="border-width:2px; border-color:black"></hr>


Exercise 2.3 (Lecture 2), Trefethen and Bau, page 15.

Let $A \in \mathbb C^{m \times m}$ be hermitian.  An eigenvector of $A$ is a nonzero vector $\mathbf x \in \mathbb C^m$ such that $A \mathbf x = \lambda \mathbf x$ for some $\lambda \in \mathbb C$, the corresponding eigenvalue.

**(a)** Prove that all eigenvalues of $A$ are real.

**(b)** Prove that if $\mathbf x$ and $\mathbf y$ are eigenvectors corresponding to distinct eigenvalues of $A$, then $\mathbf x$ and $\mathbf y$ are orthogonal.



#### Solution

$\textbf {(a)}$


Since matrix $A$ is Hermitian, then $A = A^*$.
And since the eigenvector $\mathbf x$ of matrix $A$ corresponding to eigenvalue $\lambda$ is non-zero, then we have that;
$$A\mathbf x = \lambda \mathbf x~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~(1)$$ 
Multiplying equation (1) by $\mathbf x^*$, we get;
\begin{align}
\mathbf x^* A\mathbf x &= \mathbf x^* \lambda \mathbf x \\
& = \lambda \mathbf x^* \mathbf x~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~(2) 
\end{align}
Taking the hermitian of equation (2) and using the properties $(\mathbf x^*)^* = \mathbf x$ and $(A\mathbf x)^* = \mathbf x^*A^*$, we get;
\begin{align}
(\mathbf x^* A\mathbf x)^* &= (\lambda \mathbf x^* \mathbf x)^*\\
 (A\mathbf x)^* \mathbf x &= (\mathbf x^* \mathbf x)^*\lambda ^*\\
\mathbf x^* A^* \mathbf x  &= \lambda ^* \mathbf x^* \mathbf x ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~(3)
\end{align}
But $A = A^*$, equation (3) then becomes;
\begin{align}
\mathbf x^* A \mathbf x  &= \lambda ^* \mathbf x^* \mathbf x ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~(4)
\end{align}
Equating equations (2) and (3), we get;
\begin{align}
\lambda \mathbf x^* \mathbf x  &= \lambda ^* \mathbf x^* \mathbf x\\
(\lambda -\lambda ^*)\mathbf x^* \mathbf x  &= 0
\end{align}
But $\mathbf x^* \mathbf x \neq 0$, this implies that $\lambda = \lambda ^*$.

Therefore all eigenvalues of matrix $A$ are real.


$\textbf {(b)}$

Let $\lambda _1$ and $\lambda _2$ be two distinct eigenvalues of matrix $A$ with corresponding eigenvectors $\mathbf {x}$ and $\mathbf {y}$ respectively,
we then have that;
\begin{align}
A\mathbf x &= \lambda _1 \mathbf x ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~(5) \\
A\mathbf y &= \lambda _2 \mathbf y ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~(6)
\end{align}

Multiplying equation (6) by vector $\mathbf {x^*}$ gives;

\begin{align}
\mathbf {x^*} A\mathbf y   &= \mathbf {x^*} \lambda _2 \mathbf y ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~(7)\\ 
\end{align}

Taking the hermitian of equation (7) and using the properties $(\mathbf x^*)^* = \mathbf x$, $(A\mathbf x)^* = \mathbf x^*A^*$ and that since the eigenvalues are real,$\lambda _i = \lambda _i^*$ for $i=1,2$, we get;

\begin{align}
(\mathbf {x^*} A\mathbf y)^*   &= (\mathbf {x^*} \lambda _2 \mathbf y)^* \\
(A\mathbf y)^* \mathbf x &= \lambda _2 ^* \mathbf y^* \mathbf x\\
\mathbf y^* A^* \mathbf x  &= \lambda _2 ^* \mathbf y^* \mathbf x ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~(8)
\end{align}
But $A = A^*$, equation (8) then becomes;
\begin{align}
\mathbf y^* A \mathbf x  &= \lambda _2 ^* \mathbf y^* \mathbf x ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~(9)
\end{align}
But $A \mathbf x =\lambda _1 \mathbf x  $, equation (9) then becomes;
\begin{align}
\mathbf y^* \lambda _1 \mathbf x  &= \lambda _2 ^* \mathbf y^* \mathbf x\\
\lambda _1\mathbf y^* \mathbf x &= \lambda _2 ^* \mathbf y^* \mathbf x\\
(\lambda _1 - \lambda _2 )\mathbf y^* \mathbf x &= 0
\end{align}





Since $\lambda _1$ and $\lambda _2$ are different, then $\lambda _1 - \lambda _2 \neq 0$ and this implies that $\mathbf y^* \mathbf x = 0$.

Therefore the eigenvectors $\mathbf x$ and $mathbf y$ are orthogonal. 







## Problem 4 (TB 2.4)

<hr style="border-width:2px; border-color:black"></hr>

Exercise 2.4 (Lecture 2), Trefethen and Bau, page 15-16.

What can be said about the eigenvalues of a unitary matrix? 

#### Solution

If $Q$ is unitary matrix, then $Q^{-1}  = Q^*$. For any vector $x$ we then have that;
$$~~~~$$
\begin{equation}
 \left\Vert Qx \right\Vert = \big[ (Qx)^{*} Qx \big]^{\frac{1}{2}}  =  \big[ x^*Q^* Qx \big]^{\frac{1}{2}}=  \big[ x^* I x \big]^{\frac{1}{2}} = \big[ x^*x \big]^{\frac{1}{2}} =\left\Vert x \right\Vert 
\end{equation}
$$~~~~$$
Therefore for any unitary matrix say $Q$ with an eigenvalue $\lambda$ with a corresponding eigenvector $y$ which satisfy the equation $Qy = \lambda y$, the following holds:
$$~~~~$$
\begin{equation}
 \left\Vert y \right\Vert = \left\Vert Qy \right\Vert = | \lambda| \left\Vert y \right\Vert
\end{equation}
But since $\left\Vert y \right\Vert \neq 1$, then this implies that $|\lambda| = 1$.
$$~~~~$$
Hence all eigenvalues of a unitary matrix have modulus of 1 and must lie on the unit circle in the complex plane

## Problem 5 (TB 2.5)

<hr style="border-width:2px; border-color:black"></hr>

Exercise 2.5 (Lecture 2), Trefethen and Bau, page 15-16.

Let $S \in \mathbb C^{m \times m}$ be *skew-hermitian*, i.e. $S^*=-S$. 

**(a)** Show that by using Exercise 2.3 that the eigenvalues of $S$ are pure imaginary. 

**(b)** Show that $I - S$ is non-singluar. 

**(c)** Show that the matrix $Q = (I - S)^{-1}(I + S)$, known as teh *Cayley transform of S*, is unitary.  



#### Solution

$\textbf {(a)}$


Since matrix $S$ is Hermitian, then $S = -S^*$.
Let the eigenvector $\mathbf x$ of matrix $S$ corresponding to eigenvalue $\lambda$ be non-zero, then we have that;
$$S\mathbf x = \lambda \mathbf x~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~(1)$$ 
Multiplying equation (1) by $\mathbf x^*$, we get;
\begin{align}
\mathbf x^* S\mathbf x &= \mathbf x^* \lambda \mathbf x \\
& = \lambda \mathbf x^* \mathbf x~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~(2) 
\end{align}
Taking the hermitian of equation (2) and using the properties $(\mathbf x^*)^* = \mathbf x$ and $(S\mathbf x)^* = \mathbf x^*S^*$, we get;
\begin{align}
(\mathbf x^* S\mathbf x)^* &= (\lambda \mathbf x^* \mathbf x)^*\\
 (S\mathbf x)^* \mathbf x &= (\mathbf x^* \mathbf x)^*\lambda ^*\\
\mathbf x^* S^* \mathbf x  &= \lambda ^* \mathbf x^* \mathbf x ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~(3)
\end{align}
But $S = -S^*$, equation (3) then becomes;
\begin{align}
\mathbf x^* S \mathbf x  &= -\lambda ^* \mathbf x^* \mathbf x ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~(4)
\end{align}
Equating equations (2) and (3), we get;
\begin{align}
\lambda \mathbf x^* \mathbf x  &= -\lambda ^* \mathbf x^* \mathbf x\\
(\lambda +\lambda ^*)\mathbf x^* \mathbf x  &= 0
\end{align}
But $\mathbf x^* \mathbf x \neq 0$, this implies that $\lambda = -\lambda ^*$.

Therefore the eigenvalues of  $S$ are pure imaginary since $\lambda = -\lambda ^*$.


$\textbf {(b)}$
For a matrix to be non-singular, it should be invertible implying that that its determinant is not equal to zero.
Therefore to show that matrix $I-S$ is non singular, we have to show that its eigenvaluse are non-zero.

Since $I$ is an identity matrix, its eigenvalue is 1. 

From (a) above, we have shown that since matrix $S$ is skew hermtian, then its eigenvalues $\lambda$ are purely imaginary.
The eigenvalues of matrix $I-S$ are thus given by $1-\lambda$. 

Since $1-\lambda \neq 0 $, then $det(I-S)\neq 0$ and hence matrix $I-S$ is non-singular.  




$\textbf {(c)}$


For $Q$ to be a unitary matrix, $Q^{-1} = Q^*$. This means that $QQ^{-1} =I = QQ^* $ where $I$ is an identity matrix.
Therefore to show that $Q = (I-S)^{-1}(I+S)$ is a unitary matrix, we must show that $QQ^* = I$ and this is as shown below;

\begin{align}
QQ^* &= (I-S)^{-1}(I+S)\big[(I-S)^{-1}(I+S)\big]^*\\
&= (I-S)^{-1}(I+S)(I+S)^*[(I-S)^*]^{-1}\\
&= (I-S)^{-1}(I+S)(I^*+S^*)[(I^*-S^*)]^{-1}   
\end{align}

Since $S = -S^*$ and $I^* = I$, we get;

\begin{align}
QQ^* &= (I-S)^{-1}(I+S)(I-S)[(I+S)]^{-1}   
\end{align}
But $(I+S)(I-S) = (I-S)(I+S)$, we then have;
\begin{align}
QQ^* &= (I-S)^{-1}(I-S)(I+S)[(I+S)]^{-1}
 &= I
\end{align}
Therefore since $QQ^* = I$, then $Q$ is a unitary matrix.

\begin{align}

I + \mathbf u \mathbf v^* + \alpha \mathbf u \mathbf v^* + \alpha (\mathbf v^*\mathbf u ) (\mathbf u \mathbf v^*)& = \\
 (1+\alpha+ \alpha (\mathbf v^*\mathbf u ) )\mathbf u \mathbf v^* &=0
\end{align}

If $\mathbf v^*\mathbf u = 0$, then it means that $A=I=A^{-I}$ which gives a trivial solution.

\begin{align}
 \Rightarrow 1+\alpha+ \alpha \mathbf v^*\mathbf u   &=0\\
  \Rightarrow \alpha &= \frac{-1}{1+\mathbf v^*\mathbf u}  ~~~~~~~~~~~~~and ~~~~~~~~~~~~~  \mathbf v^*\mathbf u\neq -1                           
\end{align}

If A is singular, then;

\begin{align}det(A)= det(I+\mathbf u\mathbf v^* ) &= 0\\
det(\mathbf u \mathbf v^* -(-1)I) &=0
\end{align}

This means that eigenvalues of $\mathbf u \mathbf v^*$ is -1.
Therefore

$$Au = (I+\mathbf {uv^*}) \mathbf u = I\mathbf u +\mathbf {uv^*}\mathbf u = \mathbf u-\mathbf u = 0$$

## Problem 7 (TB 3.2)

<hr style="border-width:2px; border-color:black"></hr>

Exercise 3.2 (Lecture 3), Trefethen and Bau, page 23-24.

Let $\Vert \cdot \Vert$ dnote any norm on $\mathbb C^m$ and also the induced matrix norm on $\mathbb C^{m \times m}$.  Show that  $\rho(A) \le \Vert A \Vert$, where $\rho(A)$ is the *spectral radius* fo $A$, i.e. the largest absolute value $|\lambda|$ of an eigenvalue $\lambda$ of $A$.


#### Solution
Let $\mathbf v \neq \mathbf 0$ be the eigenvector of matrix $A$ corresponding to the eigenvalue $\lambda$. We then have that;

\begin{align}
A\mathbf v &= \lambda \mathbf v\\
\Vert A\mathbf v \Vert  &= \Vert \lambda \mathbf v \Vert \\
&= |\lambda| \Vert v \Vert 
\end{align}
But $\Vert A\mathbf v \Vert \leq  \Vert A \Vert \Vert  \mathbf v \Vert$, we then get;

\begin{align}
 |\lambda| \Vert \mathbf v \Vert \leq  \Vert A \Vert \Vert  \mathbf v \Vert
\end{align}
since $\mathbf v \neq \mathbf 0$, then $\Vert \mathbf v \Vert>0$. we then have that;
\begin{align}
 |\lambda|  \leq  \Vert A \Vert 
\end{align}
Therefore for the largest absolute value $|\lambda|$ of an eigenvalue $\lambda$ of matrix $A$, we have that;
\begin{align}
\rho (A) \leq  \Vert A \Vert 
\end{align}
