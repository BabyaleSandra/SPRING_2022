### Problem 1

Theorem 5.2 in TB states that the range of  $A$ is spanned by the left singular vectors. 

Expand on the proof given in the text book by showing that $\mathbf x \in \mbox{Col}(A)$ if and only if we can write $\mathbf x$ as a linear combination of left singular vectors. 


#### Solution

Let $\mathbf {a_1,a_2,...,a_n} \in \mathbb C^m$ be columns of matrix $A\in \mathbb C^{m\times n}$ with left singular vectors $\mathbf {u_1,u_2,...,u_n}$ and right singular vectors $\mathbf {v_1,v_2,...,v_n}$  and singular values $\sigma_1, \sigma_2,...,\sigma_3$. By definition any vector say $\mathbf x \in Col(A)$ is a linear combination of columns of $A$, that is;
\begin{align}
x &= \begin{bmatrix}
| & |&\cdots & | \\ \mathbf a_1 & \mathbf a_2 & \cdots &\mathbf a_n \\ | & | &\cdots & |
\end{bmatrix} \begin{bmatrix}
w_1 \\ w_2 \\ \vdots  \\ w_n
\end{bmatrix} ~~~~~~~~~~~~~~~~~~~~~~  w_i \in \mathbb C\\
&= A\mathbf w
\end{align}

But $A = U\Sigma V^T$

\begin{align}
\Rightarrow x &=  U \Sigma V^T \mathbf w \\
& = \sum_{i=1}^{n} \sigma_i \langle \mathbf v^T_i  \; , \; \mathbf w \rangle u_i
\end{align}

Therefore $\mathbf x \in Col(A)$ iff we can write $\mathbf x$ as linear combination of left singular vectors.
